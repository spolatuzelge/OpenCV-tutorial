## GÖRÜNTÜDE SHİ-TOMAS ALGORİTMASI İLE KÖŞE TESPİTİ

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img =  cv2.imread(r"materyal\chess.png")
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(imgGray, 100, 0.01, 10) # 100 tane köşe bulacak 
corners = np.int32(corners) 

for corner in corners:
    x, y = corner.ravel()
    cv2.circle(img, (x, y), 3, (255, 0, 0), -1)

plt.imshow(img)
plt.show()
```
