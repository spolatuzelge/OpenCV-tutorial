##  GRABCUT ALGORİTMASI KULLANILARAK BİR GÖRÜNTÜDE ÖN PLAN ÇIKARIMI

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"materyal\cartoonNet.png")
mask = np.zeros(img.shape[:2],np.uint8)
mask2 = np.where((mask == 2)|(mask == 0),0,1).astype('uint8')

backGround = np.zeros((1,65),np.float64)
foreGround = np.zeros((1,65),np.float64)
```

```python
rect = (40,100,150,150)
```

```python
cv2.grabCut(img, mask, rect,
            backGround, foreGround,
            3, cv2.GC_INIT_WITH_RECT)
imgSeg = img * mask2[: , :, np.newaxis]
```

```python
plt.subplot(1,2,1)
plt.title("Orijinal")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Segmented Image')
plt.imshow(cv2.cvtColor(imgSeg, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
